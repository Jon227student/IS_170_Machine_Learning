```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df.shape
```

```python
df=pd.read_csv('/content/new data.csv', names=['sales'])
df.head() 
```

```python
data=list(df["sales"].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

df=df.replace(False,0)
df=df.replace(True,1)

df
                
```

```python
df=apriori(df,min_support=0.2,use_colnames=True)
df
```

The above values show our most frequently purchased items it seems candy is our most purchased item with 42% of our customers buying candy in their purchases. The next two are the new shooter game and the new strategy game being our top sellers, 35% of our customers include these in their purchases

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

We can note from this data that people who buy a candy usually buy a drink along with it 42% of the time also people who buy gaming laptops usually buy strategy games or rpg games 35% of the time

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

We can see a strong positive correlation between laptop purchasers and strategy games/rpg games. Console purchasers have a positive correlation for shooter games and racing games.Drink and candy also have a high correlation rate

It is also clear that laptop purchasers prefer strategy/rpg games while console purchasers enjoy shooter/racing games. We should relocate these games near eachother.

```python
df=pd.read_csv('/content/Newdata2.csv')
df.head()
```

```python
df=pd.read_csv('/content/Newdata2.csv', names=['sport goods'])
df.head() 
```

```python
df.shape
```

```python
data=list(df["sport goods"].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

df=df.replace(False,0)
df=df.replace(True,1)

df
```

```python
df=apriori(df,min_support=0.2,use_colnames=True)
df
```

It seems that head bands and water bottles are our most purchased goods here at the sporting goods store each having a .7 support rate. We can also gage what sport is selling the most product in this case tennis products have a higher frequency than the others

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

It seems that people buy head bands and water bottles most frequently together it may be because these items can go with all sports there is a 38% chance if they buy a water bottle they will buy a headband. The related sporting items also have a  antecedent of .3 for almost all sports

```python
#Let's view our interpretation values using the Association rule function.
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

Most of the related sporting items have a high lift meaning there is a positive correlation they are all over 1. Sporting items with no relation have a negative correlation most all being under 1. The average lift score for related sporting items are 3.333

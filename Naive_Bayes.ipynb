# **Naive Bayes **

The kids can play outside if the weather and temp permit. Any adversarial weather or high or low temp, the kids can't.

```python
import pandas as pd
import numpy as np

df=pd.read_csv("/content/exams.csv")
df
```

```python
#In order to process we must turn those categorical values into numberic values
#Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

#Using .fit_transform function to fit label
#encoder and return encoded label
Preparation=le.fit_transform(df['test preparation course'])

#printing weather_label
print(Preparation)
```

```python
le = preprocessing.LabelEncoder()

#Using .fit_transform function to fit label
#encoder and return encoded label
Parental=le.fit_transform(df['parental level of education'])

#printing weather_label
print(Parental)
```

```python

score_label = df['writing score'].apply(lambda x: 1 if x > 65 else 0)



score_label=le.fit_transform(df['writing_score_encoded'])
print(score_label)
```

```python
def merge (list1, list2):
  merged_list=tuple(zip(list1, list2))
  return merged_list

features=merge(Preparation, Parental)
features
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features,score_label)
#Predict Output
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
print ("Predicted Value:", predicted)
```

### From the above result, 1 indicates that the kids who did not prepare 'Failed'.

# **Naive Bayes using Multiple Labels**

```python
#import scikit-learn dataset library
from sklearn import datasets

```

```python
df=pd.read_csv("/content/babies.csv")
df.head()
```

```python
x_data = df[['gestation','age','height','bwt']]
y_data= df['smoke']
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3,random_state=109) # 70% training and 30% test
```

```python

from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Using the factors this algorithm predicts whether the mother was a smoker.
